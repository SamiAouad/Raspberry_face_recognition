```python
import cv2
import sys
from matplotlib import pyplot as plt
import uuid
import os
import requests
```

# Face detection

```python
def face_detection(img):
    cropped_faces = []
    
    face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades +"haarcascade_frontalface_default.xml")
    faces = face_cascade.detectMultiScale(img, 1.1, 4)
    for x, y, w, h in faces:
        cropped_faces.append(img[y:y+h, x:x+w])
    return cropped_faces    
```

# One face detection

```python
def surface(face_location):
    top, right, bottom, left = face_location
    return abs(bottom - top) + abs(left - right)
```

```python
def one_face_detection(img):
    cropped_faces = []
    
    face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades +"haarcascade_frontalface_default.xml")
    faces = face_cascade.detectMultiScale(img, 1.1, 4)
    if len (faces) > 0:
        closest_face_location = max(faces, key=surface)
        x, y, w, h = closest_face_location
        return img[y:y+h, x:x+w]
    return []
```

# Test

```python
PATH = "friends1.jpg"
image = cv2.imread(PATH)
faces = one_face_detection(image)
plt.imshow(cv2.cvtColor(faces, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
capture = cv2.VideoCapture(0)

while True:
    _, img = capture.read()
    cropped_faces = face_detection(img)
    print(f"the number of faces detected is: {len(cropped_faces)}")
    for cropped_face in cropped_faces:
        filename = os.path.join(os.getcwd(), 'testimages', str(uuid.uuid4()) + '.jpg')
        cv2.imwrite(filename, cropped_face)
    cv2.imshow('img', img)
    k = cv2.waitKey(30) & 0xff
    if k == 27:
        break
capture.release()
cv2.destroyAllWindows()
    
```

# Face detection using camera

```python
def record():
    capture = cv2.VideoCapture(0)
    open_door = False

    while not open_door:
        _, img = capture.read()
        cropped_face = one_face_detection(img)
        url = 'http://localhost:8000/recognize'
        myobj = {'key': cropped_face}
        _, img_encoded = cv2.imencode('.jpg', img)
        file = {'file': ('image.jpg', img_encoded.tobytes(), 'image/jpeg', {'Expires': '0'})}

        open_door = requests.post(url, files = file).json()
        cv2.imshow('img', img)
        k = cv2.waitKey(30) & 0xff
        if k == 27:
            break
    capture.release()
    cv2.destroyAllWindows()
    
```

```python
record()
```
